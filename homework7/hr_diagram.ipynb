```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt("../../course_assignments/homework7/star_data.txt",
skiprows=1)

print(data.shape) #there are 2642 rows

magnitude = data[:,0]
temperature = data[:,1]

print(np.min(magnitude)) #Main sequence like sirius
print(np.max(magnitude)) #white dwarfs

print(np.min(temperature)) #red dwarf
print(np.max(temperature)) #white dwarfs, MS, and supergiants


#6. the minimum magnitude star is the brightest and the maximum magnitude star is the dimmest
```

```python
log_magnitude= np.log(magnitude)
log_temperature= np.log(temperature)

fig= plt.figure(figsize=(8,6))
plt.scatter(log_temperature, log_magnitude)

plt.gca().invert_xaxis()
plt.gca().invert_yaxis()

Brightest_star= np.argmin(magnitude)
Dimmest_star= np.argmax(magnitude)
Hottest_star= np.argmax(temperature)
Coldest_star= np.argmin(temperature)

plt.scatter(log_temperature[Brightest_star], log_magnitude[Brightest_star], color='black', marker = '*', s = 300, label='Brightest_star')
plt.scatter(log_temperature[Dimmest_star], log_magnitude[Dimmest_star], color='purple', marker = '*', s = 300, label='Dimmest_star')
plt.scatter(log_temperature[Hottest_star], log_magnitude[Hottest_star], color='green', marker = '*', s = 300, label='Hottest_star')
plt.scatter(log_temperature[Coldest_star], log_magnitude[Coldest_star], color='orange', marker = '*', s = 300, label='Coldest_star')

plt.legend()
plt.title('Hertzsprung-Russel Diagram')
plt.xlabel('Temperature (log K)')
plt.ylabel('Magnitude')
```
