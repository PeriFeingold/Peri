```python
#2 Matplotlib
import numpy as np
import matplotlib.pyplot as plt

def make_sine_wave(x,A,w):
    return A*np.sin(w*x)

x=np.linspace(0,2 * np.pi, 1000)

amplitude = [0.5, 1, 1.5, 2, 2.5]
frequency = [1, 2, 3, 4, 5]

fig, ax=plt.subplots(1,3, figsize=(25,5))

first_subplot= ax[0]
second_subplot= ax[1]
third_subplot= ax[2]
zip(amplitude, frequency)

#1st sublot
for A in amplitude:
    y= make_sine_wave(x,A,1)
    ax[0].plot(x, y, label=f"A={A}, w=1")

#2nd sublot
for w in frequency:
    y= make_sine_wave(x,1,w)
    ax[1].plot(x, y, label=f"A=1, w={w}")

#3rd subplot
for A,w, in zip(amplitude, frequency):
    y= make_sine_wave(x,A,w)
    ax[2].plot(x, y, label=f"A={A}, w={w}")

ax[0].set_title("Varying Amplitude")
ax[1].set_title("Varying Frequency")
ax[2].set_title("Varying Amplitude and Frequency")
plt.show()
```
