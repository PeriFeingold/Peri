```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#3.1 Plotting an Astronomical Image
img = np.loadtxt("astro_image.csv")
plt.figure(figsize=(8, 6))
plt.imshow(img)
plt.colorbar()
plt.title("Astronomy Image")
plt.show()
```

```python
zoom1 = img[1000:1200, 800:1000]  
zoom2 = img[1200:1400, 800:1000]

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.imshow(zoom1)
plt.title("Zoom 1")
plt.colorbar()

plt.subplot(1, 2, 2)
plt.imshow(zoom2)
plt.title("Zoom 2")
plt.colorbar()

plt.tight_layout()
plt.show()
```
