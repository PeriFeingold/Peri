```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
```

```python
# 3.1 read FITS files
fits_files = [
    "M13_0001.fits",
    "M13_0002.fits",
    "M13_0003.fits",
    "M13_0004.fits",
    "M13_0005.fits"
]

fits_data = []

# Loop through each FITS file
for file in fits_files:
    with fits.open(file) as hdul:
        data = hdul[0].data    # Get the image data
        fits_data.append(data) # Add it to our list

#3.2 plot the images
fig, axes = plt.subplots(1, 5, figsize=(20, 4))  # 1 row, 5 columns

for i, ax in enumerate(axes):
    im = ax.imshow(fits_data[i], origin='lower')  
    ax.set_title(f"M13 0000{i+1}")
   

plt.suptitle("FITS Images")
plt.show()
```

```python
# second set of 1x5 subplots
all_data = np.array(fits_data)
vmin = np.min(all_data)
vmax = np.max(all_data)

fig, axes = plt.subplots(1, 5, figsize=(20, 4))

for i, ax in enumerate(axes):
    im = ax.imshow(fits_data[i], origin='lower', vmin=vmin, vmax=vmax, cmap='magma')
    ax.set_title(f"M13 0000{i+1}")
    ax.axis('off')

cbar = fig.colorbar(im, ax=axes.ravel().tolist(), shrink=0.6)
cbar.set_label("Pixel Value")

plt.suptitle("FITS Images - Same Color Scale")
plt.show()
```

```python
# I noticed the images are a lot harder to see
```

```python
# 3.3 Plot Histograms
fig, axes = plt.subplots(1, 5, figsize=(20, 4))

for i, ax in enumerate(axes):
    ax.hist(fits_data[i].ravel(), bins=50, color='blue', alpha=0.7)
    ax.set_title(f"M13 0000{i+1}")
    ax.set_xlabel("Pixel Value")
    ax.set_ylabel("Number of Pixels")

plt.suptitle("Histograms of FITS Images")
plt.show()
```

```python
# 2
# a) x-axis = pixel value and the y-axis =number of pixels
# b) bin=50 means the pixel value range is divided into 50 intervals
# c) around 500
# d) the data is not very spread out
# e) mean and median value for each image
for i, data in enumerate(fits_data):
    median_val = np.median(data)
    mean_val = np.mean(data)
    print(f"M13 0000{i+1}: median = {median_val:.2f}, mean = {mean_val:.2f}")
```

```python
# 3.4 Combine and Normalize Images
combined_image = np.median(fits_data, axis=0)
combined_median = np.median(combined_image)
print(f"Combined image median = {combined_median:.2f}")
```

```python
plt.hist(combined_image.ravel(), bins=50, alpha=0.7)
plt.xlabel("Pixel Value")
plt.ylabel("Number of Pixels")
plt.title("Histogram of Combined Image")
plt.show()
```

```python
plt.imshow(combined_image, origin='lower', vmin=450, vmax=600)
plt.colorbar(label='Pixel Value')
plt.title("Combined M13 Image")
plt.show()
```
