```python
# apply astropy units

def attach_units(col, unit, new_col=None):
    new_values = []
    for val in df[col]:
        new_values.append(val * unit)
    df[col] = new_values
    if new_col is not None:
        df.rename(columns={col: new_col}, inplace=True)

attach_units(
    col='Perihelion (km)',
    unit=10**6 * u.km,
    new_col="Perihelion (km)"
)

print(df["Perihelion (km)"])
```

```python
# 2.3 data analysis

# have to make Aphelion the same units first
def attach_units(col, unit, new_col=None):
    new_values = []
    for val in df[col]:
        new_values.append(val * unit)
    df[col] = new_values
    if new_col is not None:
        df.rename(columns={col: new_col}, inplace=True)
attach_units(
    col="Aphelion (km)",
    unit=10**6 * u.km,
    new_col="Aphelion (km)"
)

print(df["Aphelion (km)"])
```

```python
# 1. semi-major axis

semi_major = (df["Perihelion (km)"] + df["Aphelion (km)"]) / 2
aphelion_index = df.columns.get_loc("Aphelion (km)")
df.insert(aphelion_index + 1, "Semi-Major Axis (km)", semi_major)

print(df.columns)
print(df["Semi-Major Axis (km)"])
```

```python
# Convert Orbital Period to Years
df["Orbital Period (yr)"] = df["Orbital Period (days)"] * u.day.to(u.yr)
df.rename(columns={"Orbital Period (days)": "Orbital Period (yr)"}, inplace=True)
print(df.columns)
print(df["Orbital Period (yr)"])
earth_period = df.loc[df["Planet"] == "Earth", "Orbital Period (yr)"].iloc[0]
print(f"Earth’s orbital period: {earth_period:.4f}")
```

```python
# Convert Distance Columns
# AU stands for astronomical units and its a measurment for the distance between earth and the sun.
const.au
```

```python
print(const.au.to(u.km))
```

```python
distance_cols = [
    "Diameter (AU)",
    "Perihelion (AU)",
    "Aphelion (AU)",
    "Semi-Major Axis (AU)",
] #for it to run i need to manually type AU

for col in distance_cols:
    df[col] = df[col] * u.km.to(u.AU)

for col in distance_cols:
    new_col = col.replace("(km)", "(AU)")
    df.rename(columns={col: new_col}, inplace=True)

planet = "Earth"   # change to your favorite
for col in df.columns:
    if "(AU)" in col:
        value = df.loc[df["Planet"] == planet, col].iloc[0]
        print(f"{planet} — {col}: {value}")
```
