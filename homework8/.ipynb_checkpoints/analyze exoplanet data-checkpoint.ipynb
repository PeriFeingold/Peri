```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('exoplanet_data.csv', comment="#")
print(df)
```

```python
#4.
# the shape is 570 by 92
# there are 570 rows and 92 columns
```

```python
# print all the column titles
print(df.columns)
```

```python
# List 5 column titles that you have no idea what they mean.
#  pl_orbeccenerr2, pl_orbeccenlim, pl_insolerr2, rastr, ttv_flag


#3.
# table label: Planet Name
    # Database Column Name: pl_name
    # description of the parameter: the common name of a planet

# table label: Host Name
    # Database Column Name: hostname
    # description of the parameter: the common name for the host star

# table label: Discovery Method
    # Database Column Name: discoverymethod
    # description of the parameter: the method of how the planet was found

# table label: Spectral Type
    # Database Column Name: st_spectype	
    # description of the parameter: the classification of a star based on its spectral type/characteristics

# table label: Equilibrium Temperature [K]
    # Database Column Name: pl_eqt
    # description of the parameter: the equilibrium temperature of a planet heated by its host star.

# table label: Planet Mass or Mass*sin(i) [Jupiter Mass]
    # Database Column Name: pl_bmassj	
    # description of the parameter: the best estimate for mass of a planet. Measured in jupiter masses. 
```

```python
smaller_df= df[['pl_name','hostname','discoverymethod', 'st_spectype', 'pl_eqt', 'pl_bmassj']]
print(smaller_df)
```

```python
#5. filter smaller data set
new_df=smaller_df.dropna()
print(new_df)
```

```python
# the shape if 67 by 6
# there are 67 rows and 6 columns
```

```python
#print the first 5 rows
new_df.head()
```

```python
#print the last 5 rows
new_df.tail()
```

```python
#Print all the column titles.
print(new_df.columns)
```

```python
new_df.value_counts('discoverymethod')
#the different methods are transit, Radial Velocity, and Imaging
# there are 52 transits, 13 radial velocities, and 2 imaging. Transit has the most and imaging has the least
# transit: when the brightness of a star dips because an exoplanet passed infront of it
# radial velocity: detecting the doppler effect in a star's motion caused by the gravity of an orbiting planet
# imaging: taking a picture of the exoplanet
```

```python
plt.figure(figsize=(8,6))
rv = new_df[new_df['discoverymethod'] == 'Radial Velocity']
plt.scatter(x=(rv['pl_bmassj']),y=(rv['pl_eqt']), color='blue', label='Radial Velocity')

T = new_df[new_df['discoverymethod'] == 'Transit']
plt.scatter(x=(T['pl_bmassj']),y=(T['pl_eqt']), color='red', label='Transit')

img = new_df[new_df['discoverymethod'] == 'Imaging']
plt.scatter(x=(img['pl_bmassj']),y=(img['pl_eqt']), color='yellow', label='Imaging')

plt.title("Planet Mass vs. Equilibrium Temperature")
plt.xlabel("Planet Mass (Earth Masses)")
plt.ylabel("Equilibrium Temperature (K)")
plt.legend()
plt.show()
```

```python
#Duplicate your previous plot but change both axes to a logarithmic scale.

plt.figure(figsize=(8,6))
rv = new_df[new_df['discoverymethod'] == 'Radial Velocity']
plt.scatter(x=np.log(rv['pl_bmassj']),y=np.log(rv['pl_eqt']), color='blue', label='Radial Velocity')

T = new_df[new_df['discoverymethod'] == 'Transit']
plt.scatter(x=np.log(T['pl_bmassj']),y=np.log(T['pl_eqt']), color='red', label='Transit')

img = new_df[new_df['discoverymethod'] == 'Imaging']
plt.scatter(x=np.log(img['pl_bmassj']),y=np.log(img['pl_eqt']), color='yellow', label='Imaging')

plt.title("Planet Mass vs. Equilibrium Temperature")
plt.xlabel("Planet Mass (Earth Masses)")
plt.ylabel("Equilibrium Temperature (K)")
plt.legend()
plt.show()
```

```python
#4. based on the plot, exoplanets found by imaging are more likely to be massive. 
```

```python
massive_planet=new_df[new_df['pl_bmassj']>5]
coldest_and_massive_planet=massive_planet.loc[massive_planet['pl_eqt'].idxmin()]
print(coldest_and_massive_planet)
```

```python
plt.figure(figsize=(8,6))
plt.figure(figsize=(8,6))
rv = new_df[new_df['discoverymethod'] == 'Radial Velocity']
plt.scatter(x=np.log(rv['pl_bmassj']),y=np.log(rv['pl_eqt']), color='blue', label='Radial Velocity')

T = new_df[new_df['discoverymethod'] == 'Transit']
plt.scatter(x=np.log(T['pl_bmassj']),y=np.log(T['pl_eqt']), color='red', label='Transit')

img = new_df[new_df['discoverymethod'] == 'Imaging']
plt.scatter(x=np.log(img['pl_bmassj']),y=np.log(img['pl_eqt']), color='yellow', label='Imaging')
plt.scatter(x=np.log(coldest_and_massive_planet['pl_bmassj']),y=np.log(coldest_and_massive_planet['pl_eqt']), color='pink', label= 'coldest_and_massive_planet', s=200)

plt.title("Planet Mass vs. Equilibrium Temperature")
plt.xlabel("Planet Mass (Earth Masses)")
plt.ylabel("Equilibrium Temperature (K)")
plt.legend()
plt.show()
```
