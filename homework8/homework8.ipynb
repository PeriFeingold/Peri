```python
median_latitude=df['latitude'].median()
df['absolute_distance_to_median_latitude']= (df['latitude']-median_latitude).abs()
df.head()
#I dont know why its making two coloumns
```

```python
median_longitude=df['longitude'].median()
df['absolute_distance_to_median_longitude']= (df['longitude']-median_longitude).abs()
df.head()
```

```python
df.shape
#52 rows and 7 columns
```

```python
closest_latitude=df.loc[df['absolute_distance_to_median_latitude'].idxmin()]
print(closest_latitude)
print(f"{closest_latitude['absolute_distance_to_median_latitude']:.2f}")
```

```python
farthest_latitude=df.loc[df['absolute_distance_to_median_latitude'].idxmax()]
print(farthest_latitude)
print(f"{farthest_latitude['absolute_distance_to_median_latitude']:.2f}")
```

```python
closest_longitude=df.loc[df['absolute_distance_to_median_longitude'].idxmin()]
print(closest_longitude)
print(f"{closest_longitude['absolute_distance_to_median_longitude']:.2f}")
```

```python
farthest_longitude=df.loc[df['absolute_distance_to_median_longitude'].idxmax()]
print(farthest_longitude)
print(f"{farthest_longitude['absolute_distance_to_median_longitude']:.2f}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df= sns.load_dataset('taxis')
print(df)
df.shape
```

```python
#getting rid of NaN values
df.dropna(axis=1, how='any')
```

```python
# the shape is (6433, 9)
# there are 6433 rows and 9 columns
```

```python
#print the first 5 rows
df.head()
```

```python
# date mean the date recorded 
# extent means the amount of area covered by ice
```

```python
fig, ax=plt.subplots(2,2, figsize=(20,20))

sns.scatterplot(data=df, x='pickup', y='dropoff', color='blue', ax=ax[0,0], label='taxis')
ax[0,0].set_title('Pickup vs Dropoff')
ax[0, 0].set_xlabel("Pickup (location)")
ax[0, 0].set_ylabel("Dropoff (location)")
ax[0,0].legend(loc='upper left') #idk why its not working come back to this

sns.scatterplot(data=df, x='fare', y='distance', color='red', ax=ax[0,1])
ax[0,1].set_title('Distance vs Fare')
ax[0, 1].set_xlabel("Distance (miles)")
ax[0, 1].set_ylabel("Fare ($)")

sns.histplot(data=df, x='tip', color='green', ax=ax[1,0])
ax[1,0].set_title('Amount Tipped')
ax[1,0].set_xlabel("tip ($)")


sns.boxplot(data=df, y='passengers', color='purple', ax=ax[1,1])
ax[1,1].set_title('Distribution of Passangers')

plt.show()

```

```python
# the data in subplot [0,0] is the plot of pickup vs dropoff location. I picked these columns because they were the first two.
# the data in subplot [0,1] is the plot of distance vs how much a person paid to go that distance. I picked this because I wanted to see the varying prices.
# the data in subplot [1,0] is the distribution of tips. It shows that the majority of people didn't tip. I picked this because I wanted to see if people were kind and tipped. 
# the data in subplot [1,1] is the distribution of the amount of passangers per trip. It tells us the median was 1-2 passangers. I picked this column to see what middle amount of people that rode in a taxi at a time.
```
