```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
df = pd.read_csv("noisy_curvefit_data.csv")
print(df)
```

```python
#4.
#the shape is 100x6
#there are 100 rows and 6 columns
#print the first 5 rows and the last 5 rows
df.head()
df.tail()
```

```python
#the column names
print(df.columns)
```

```python
x=df['x']
y=df['y_noise_1']

plt.scatter(x,y, color= 'gray', label='Noisy data')
plt.xlabel('x')
plt.ylabel('first noise level')
plt.title('x vs. y_noise_1')
plt.legend()
plt.show()
```

```python
#Fit the Data
def model(x, A, k):
    return A * np.exp(-k * x)

p0 = [1, 1]
popt, pcov = curve_fit(model, x, y, p0=p0) #y was defined as y_noise_1 in the problem above
A_fit, k_fit = popt
print(f"Fit: y = {A_fit:.2f} * e^(-{k_fit:.2f} * x)")
```

```python
residuals = y - model(x, A_fit, k_fit)
chi2 = np.sum(residuals**2)
dof = len(x) - len(popt)
chi2_reduced = chi2 / dof
print(f"Reduced chi-squared: {chi2_reduced:.2f}")
```

```python
plt.scatter(x,y, color= 'gray', label='Noisy data')
plt.xlabel('x')
plt.ylabel('y_noise_1')
plt.title('First Noise Level with 0.79 Reduced chi-squared')

plt.plot(popt, color='red', label='fitted model')

plt.plot(x, model(x, 5, 0.8), color='black', linestyle='--', label='true model')


plt.legend()
plt.show()
```

```python
#Compare different noise levels

def fit_model(x,y):
    p0 = [1, 1]
    popt, pcov = curve_fit(model, x, y, p0=p0)
    A_fit, k_fit = popt
    residuals = y - model(x, A_fit, k_fit)
    chi2 = np.sum(residuals**2)
    dof = len(x) - len(popt)
    chi2_reduced = chi2 / dof
    return popt,  chi2_reduced

fig, axes=plt.subplots(2,2, figsize=(10,8))
axes = axes.flatten()

noise_columns = ['y_noise_2', 'y_noise_3', 'y_noise_4', 'y_noise_5']
x=df['x']

for i, (ax, y_noise)  in enumerate(zip(axes, noise_columns), start=2):
    p0 = [1, 1]
    popt, pcov = curve_fit(model, x, y, p0=p0)
    A_fit, k_fit = popt
    print(f"Fit: y = {A_fit:.2f} * e^(-{k_fit:.2f} * x)")
    
    ax.scatter(x, y = df[y_noise], color= 'gray', label='Noisy data')
    ax.plot(x, model(x, A_fit, k_fit), color='red', label='fitted model')
    ax.plot(x, model(x, 5, 0.8), color='black', linestyle='--', label='true model')

    ax.set_title(f"Noise level {noise_level}\nReduced Chi = {chi2_red:.3f}")
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.legend()
    
plt.show()
df['y_noise_2']
```

```python
#the red and black lines get less similar as you increase noise level
#reduced chi-squared gets bigger as u 
```

```python
#new pandas dataframe

results = []
noise_levels = [0, 2, 3, 4, 5]
x = df['x']

for nl in noise_levels:
    y = df[f'y_noise_{nl}'] if nl != 0 else df['y_noise_1'] 
    popt, chi2_red = fit_model(x, y)
    A_fit, k_fit = popt
    results.append({
        "Noise level": nl,
        "A_fit": A_fit,
        "k_fit": k_fit,
        "Reduced chi-squared": chi2_red
    })
noisy_parameters= pd.DataFrame(results)
noisy_parameters

df.to_csv("noisy_parameters.csv", index=False)
```
